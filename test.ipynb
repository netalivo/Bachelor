```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten einlesen und vorbereiten
df = pd.read_csv('Walmart_Sales.csv', parse_dates=['Date'], dayfirst=True)
df.columns = df.columns.str.lower()
df.set_index('date', inplace=True)

# Nur die ausgewhlten Stores plotten
selected_stores = [17, 30, 37, 44]

for store_number in selected_stores:
    sales = df[df['store'] == store_number]['weekly_sales'].asfreq('W-FRI')
    plt.figure(figsize=(10, 4))
    plt.plot(sales, marker='o', linestyle='-')
    plt.title(f'Weekly Sales fr Store {store_number}')
    plt.xlabel('Datum')
    plt.ylabel('Weekly Sales')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

```
